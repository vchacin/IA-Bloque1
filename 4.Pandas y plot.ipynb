## Histogramas

El histograma es útil para observar la desidad de los valores. Por ejemplo, es posible que desee saber cuántos niveles de potencia son mayores que 90. El histograma nos permite ver la distribución de probabilidad de nuestras variables.

Veamos cómo se trazan los histogramas.

```python
y = dataframe.Power

plt.hist(y,bins=10)

plt.show()
```

¿Qué significa lo anterior?

Etiquetemos el gráfico claramente de la siguiente manera:

- Título: Distribución de probabilidad de niveles de potencia
- Eje Y: Número de Pokémon
- Eje X: Nivel de potencia

```python
y = dataframe.Power

plt.hist(y,bins=10)

plt.ylabel("Cantidad de Pokémones")
plt.xlabel("Nivel de potencia")
plt.title('Distribución de probabilidad de niveles de potencia')

plt.show()
```

### Tarea 8: Trace un gráfico de barras con los marcadores del eje x girados a 90 grados.

```python
objects = dataframe.Type1.unique()
values = dataframe['Type1'].value_counts()
y_pos = np.arange(len(objects))
plt.bar(y_pos, values, align='center')
plt.xticks(y_pos, values.index) 
plt.xticks (rotation=45)
plt.show()
```

### Tarea 9: Agregar un título al eje X y otro al eje Y

```python
objects = dataframe.Type1.unique()
values = dataframe['Type1'].value_counts()
y_pos = np.arange(len(objects))
plt.bar(y_pos, values, align='center')
plt.xticks(y_pos, values.index) # name the xticks
plt.xticks (rotation=60)
# Add x and y labels and set a font size
plt.xlabel ("Type", fontsize = 14)
plt.ylabel ("Counts", fontsize = 14)
plt.show()
```

### Tarea 13: Cambiar el color de fuente de las barras en el gráfico de barras

```python
plt.figure(figsize=(10,10))
objects = dataframe.Type1.unique()
values = dataframe['Type1'].value_counts()
y_pos = np.arange(len(objects))
plt.bar(y_pos, values, align='center', color="orange")
plt.xticks(y_pos, values.index) # name the xticks
plt.xticks (rotation=60, color="green")
# Add x and y labels and set a font size
plt.xlabel ("Pokemon Type", fontsize = 14)
plt.title("Pokemon Type Frequency", fontsize = 25)
plt.ylabel ("Frequencies", fontsize = 14)
plt.show()
```

El siguiente tema interesante que veremos es el gráfico de doble barra. En este ejemplo, habrá 2 barras por cada valor de x en el gráfico de barras. Eche un vistazo al código a continuación e intente completarlo. Intentaremos trazar las frecuencias de las variables Tipo1 y Tipo2 en el conjunto de datos.

### Tarea 14: Crea un gráfico de barras dobles usando el código siguiente.

```python
plt.figure(figsize=(20,10))
objects = dataframe.Type1.unique()
values = dataframe['Type1'].value_counts()
values_1 = dataframe['Type2'].value_counts()
y_pos = np.arange(len(objects))
ax = plt.subplot(111)
ax.bar(y_pos, values,width = 0.3, align='center',color='black')
ax.bar(y_pos-0.3, values_1,width = 0.3, align='center')

plt.xticks(y_pos, values.index) # name the xticks
plt.xticks (rotation=60)
# Add x and y labels and set a font size
plt.xlabel ("Pokemon Type", fontsize = 14)
plt.title("Pokemon Type Frequency", fontsize = 25)
plt.ylabel ("Frequencies", fontsize = 14)
plt.show()
```

## Visualización de imágenes

La visualización de imágenes es una parte importante para comprender el conjunto de datos. La mayoría de los conjuntos de datos de visión por computadora vienen con imágenes y algunos datos de respaldo, como puede ver en el archivo pokemon.csv. Ahora que hemos terminado de explorar el archivo CSV, comenzaremos a explorar las imágenes de Pokémon.

Aquí estamos tratando de imprimir una imagen usando la biblioteca matplotlib. Pruebe esto también con otras imágenes. ¿Encuentra algo interesante?

```python
img = mpimg.imread('PokeIMG/images/abra.png')
plt.imshow(img)
plt.show()
```

Hemos entendido cómo las imágenes se componen de 3 canales. ¿se ha preguntado cómo podemos trazar un solo canal de la imagen? Podemos crear subconjuntos de los datos de la imagen y trazarlos. Para el siguiente ejemplo, solo estamos considerando el canal verde de una imagen, etiquetado como canal # 1.

También usaremos el atributo cmap de la función imshow para trazar la imagen como un mapa de calor. Puede obtener más información sobre él [aquí] (https://stackoverflow.com/questions/25625952/matplotlib-what-is-the-function-of-cmap-in-imshow).

```python
img = mpimg.imread('PokeIMG/images/abra.png')
img = img[:,:,1]
plt.imshow(img, cmap='hot')
plt.show()
```

El próximo desafío es mostrar una barra de color junto a la imagen que estamos considerando hoy. Utilice la función de barra de colores sobre la que puede obtener más información [aquí] (https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.colorbar.html).

### Tarea 15: Agrega una barra de color al gráfico

```python
img = mpimg.imread('PokeIMG/images/abra.png')
img = img[:,:,1]
plt.imshow(img, cmap='hot')
plt.colorbar()
plt.show()
```

### Tarea 16: Grafica 2 imágenes en la misma parcela usando subparcelas
Las subparcelas son herramientas útiles para crear múltiples parcelas en una sola imagen. Consulte el [enlace] (https://stackoverflow.com/questions/3584805/in-matplotlib-what-does-the-argument-mean-in-fig-add-subplot111) para obtener información adicional.

```python
plt.figure(figsize=(10,10))
plt.subplot(211)
img = mpimg.imread('PokeIMG/images/abra.png')
plt.imshow(img)
plt.subplot(212)
img = mpimg.imread('PokeIMG/images/charmander.png')
plt.imshow(img)
plt.show()
```
